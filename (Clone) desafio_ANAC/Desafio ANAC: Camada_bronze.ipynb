### Desafio de arquitetura medalhao, onde sera feita a limpeza dos dados da ANAC

**1.Ler os dados da ANAC que estao em csv e compacta-los em parquet**

```python
#fazendo a leitura dos dados
#caminho salvo em uma var
df_anac_bronze_origem = (
    "/Volumes/pratica/pratica_dbs/praticas_vol/Anac/V_OCORRENCIA_AMPLA.csv"
)

df_anac_dados_bronze = (
    spark.read.format("csv")
    .options(header=True, sep=";", inferSchema=True, skipRows=1)
    .load(df_anac_bronze_origem)
)

display(df_anac_dados_bronze.limit(10))
```

```python
# Bronze/anac_raw salvar com esse nome "anac_raw_bronze"
# O salvamento deve ser feito no formato Parquet e com a compressão GZIP (para economizar espaço de armazenamento bruto).
df_anac_bronze_caminho = "/Volumes/pratica/pratica_dbs/bronze/anac_raw_bronze"

(
    df_anac_dados_bronze.write.format("parquet")
    .options(compression="gzip")
    .mode("overwrite")
    .save(df_anac_bronze_caminho)
)

#parquet nao precisa de options, como sep, header ou inferschema o seu salvamento
#batsa apenas passar o formato e salvar e mode de sobrescrita

#Correcoes feitas, remocao das options (header=True, sep=";", infercSchema=True, skipRows=1)
```
